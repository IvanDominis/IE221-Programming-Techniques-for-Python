1. Xây dựng lớp khái niệm điểm và các phương thức thiết lập, nhập, xuất điểm,
thay đổi tọa độ của điểm, lấy hoành độ, tung độ, tịnh tiến.

```python
class diem():
  def _init_(self,x,y):
    self.x = x 
    self.y = y
  def nhap(self):
    self.x, self.y = map(int,input('Nhập điểm: ').split())
  def xuat(self):
    print('Điểm vừa nhập là: ' + '('+str(self.x)+','+str(self.y)+')')
  def doi(self):
    a,b = map(int,input('Nhập tọa độ thay thế: ').split())
    self.x = a
    self.x = b
  def tung(self):
    return self.x
  def hoanh(self):
    return self.y
  def tinhtien(self):
    a = diem()
    print('Điểm tịnh tiến')
    a.nhap()
    self.x += a.x
    self.y += a.y
```

NOTE: Nhập điểm có dạng a b lần lượt là tung độ và hoành độ của điểm
Ví dụ: (2,3) -> 2 3

```python
a = diem()
a.nhap()
a.xuat()
a.doi()
print('Hoành độ của điểm vừa nhập là:',a.hoanh())
print('Hoành độ của điểm vừa nhập là:',a.tung())
a.tinhtien()
a.xuat()
```

2. Xây dựng lớp TamGiac để biểu diễn khái niệm tam giác trong mặt phẳng với
các phương thức thiết lập, nhập, xuất, tịnh tiến, lấy trọng tâm tam giác.

```python
def cong(a,b):
  t = diem()
  t.x = a.x + b.x
  t.y = a.y + b.y
  return t
```

```python
class TamGiac():
  def _init_(self,a,b,c,g):
    self.a = a
    self.b = b
    self.c = c
    self.g = g
  def nhap(self):
    l = []
    gx, gy = 0, 0
    print('Nhập 3 điểm của tam giác')
    for i in range(3):
        temp = diem()
        temp.nhap()
        gx +=temp.x
        gy +=temp.y
        l.append(temp)
    self.a = l[0]
    self.b = l[1]
    self.c = l[2]
    self.g = diem()
    self.g.x = gx/3
    self.g.y = gx/3 
  def xuat(self):
    print('Tọa độ 3 điểm của tam giác là:')
    self.a.xuat()
    self.b.xuat()
    self.c.xuat()
  def TrongTam(self):
    return self.g
  def TinhTien(self):
    print('Nhập điểm tịnh tiến')
    t = diem()
    t.nhap()
    self.a = cong(self.a,t)
    self.b = cong(self.b,t)
    self.c = cong(self.c,t)
```

```python
b = TamGiac()
b.nhap()
b.xuat()
g = b.TrongTam()
print('Trọng tâm của tam giác vừa nhâp là: ('+str(g.x)+','+str(g.y)+')')
b.TinhTien()
b.xuat()
```

```python
import numpy as np
class Matrix():
  # 1. Phương thức khởi tạo dùng để khởi tạo ma trận có số dòng và số cột được truyền vào.
  def _init_(self,rows,columns,elements):
    self.rows = rows
    self.cols = columns
    self.elements = elements
  # 2. Phương thức getRows dùng để lấy ra số lượng dòng của ma trận.
  def getRows(self):
    return self.rows
  # 3. Phương thức getCols dùng để lấy ra số lượng cột của ma trận.
  def getCols(self):
    return self.cols
  # 4. Phương thức write dùng để nhập vào các phần tử của ma trận
  def write(self):
    l = []
    print('----Nhập phần tử của ma trận----')
    self.rows = int(input('Nhập số dòng của ma trận: '))
    for i in range(self.rows):
      n = list(map(int,input().split()))
      l.append(n)
    self.elements = l
    self.cols = len(l[0])
  # 5. Phương thức getEle dùng để lấy ra phần tử của ma trận tại vị trí (i, j).
  def getEle(self,i,j):
    return self.elements[i-1][j-1]
  # 6. Phương thức add dùng để cộng hai ma trận được truyền vào
  def add(self,a):
    t = []
    if self.cols == a.cols and self.rows == a.rows:
      b = Matrix()
      for i in range(self.rows):
        c = []
        for j in range(self.cols):
          c.append(self.elements[i][j]+a.elements[i][j])
        t.append(c)
      return t
    return None
  # 7. Phương thức subtract dùng để trừ hai ma trận được truyền vào
  def subtract(self,a):
    t = []
    if self.cols == a.cols and self.rows == a.rows:
      b = Matrix()
      for i in range(self.rows):
        c = []
        for j in range(self.cols):
          c.append(self.elements[i][j]-a.elements[i][j])
        t.append(c)
      return t
    return None
  # 8. Phương thức multiply dùng để nhân hai ma trận được truyền vào
  def multiply(self,a):
    t = []
    if self.cols == a.cols and self.rows == a.rows:
      for i in range(self.rows):
        c = []
        for j in range(self.cols):
          c.append(self.elements[i][j]*a.elements[i][j])
        t.append(c)
      return t
    return None
  # 9.Phương thức inverse dùng để tìm ma trận nghịch đảo của ma trận được truyền vào
  def inverse(self):
    if self.rows == self.cols:
      return np.linalg.inv(self.elements)
    return None
```

Khai báo lớp đối tượng với các thuộc tính và phương thức tương ứng.

```python
class SSeven():
  def _init_(self,a,b):
    self.a = a
    self.b = b
  def nhap(self):
    self.a = int(input('Nhập a: '))
    self.b = int(input('Nhập b: '))
  def xuat(self):
    print('Xuất:')
    if self.b == 0:
      print(str(self.a))
    elif self.b > 0:
      print(str(self.a) + ' + ' + str(self.b) + '√7')
    else:
      print(str(self.a) + ' - ' + str(-self.b) + '√7')
  def cong(self,x):
    t = SSeven()
    t.a = self.a + x.a
    t.b = self.b + x.b
    return t
  def tru(self,x):
    t = SSeven()
    t.a = self.a - x.a
    t.b = self.b - x.b
    return t
  def nhan(self,x):
    t = SSeven()
    t.a = self.a * x.a
    t.b = self.b * x.b
    return t
  def chia(self,x):
    t = SSeven()
    t.a = self.a / x.a
    t.b = self.b / x.b
    return t
  def doidau(self):
    self.a = -self.a
    self.b = -self.b
```

# Bài 22

Viết chương trình cho phép thực hiện các chức năng sau:
1. Nhập vào danh sách các vé.
2. Tính tổng tiền vé mà công viên thu được.
3. Cho biết có bao nhiêu vé đã bán là vé từng phần.

```python
class ticket():
  def _init_(self,id,name,dob,games):
    self.id = id
    self.name = name
    self.dob = dob
    self.games = games
  def input(self):
    self.id = int(input('Please enter the id of ticket: '))
    self.name = input('Enter name: ')
    self.dob = input('Enter date of birth: ')

class PackageTicket(ticket):
  def _init_(self):
    ticket._init_(self)
  def Calc_price(self):
    return 200000

class PartiallyTicket(ticket):
  def _init_(self):
    ticket._init_(self)
  def input(self):
    ticket.input(self)
    self.games = int(input('How many games do you want to play: '))
  def Calc_price(self):
    return 70000 + 20000*self.games 
```

```python
l = []
num = [0,0]
sum_price = 0
n = int(input('Enter the number of guest: '))
print('-----Provide the information of tickets-----')
print('There are two kinds of ticket: 0 - Package Ticket, 1 - Partially Ticket')
print('Please choose 0 or 1 to select type of ticket!')
for i in range(n):
  k = int(input('Select ticket: '))
  if k == 0:
    t = PackageTicket()
    t.input()
    sum_price += t.Calc_price()
    num[0] +=1
    l.append(t)
  elif k == 1:
    t = PartiallyTicket()
    t.input()
    sum_price += t.Calc_price()
    num[1] +=1
    l.append(t)

print('Total money earned from selling ticket: ' + str(sum_price) + 'VND')
print('There are ' + str(num[1]) +' partially tickets are consumed')
```

# Bài 24

Áp dụng kiến thức về kế thừa và đa hình, hãy vẽ sơ đồ lớp và viết chương trình
thực hiện các chức năng sau:
1. Nhập danh sách N nhân viên của công ty (N được nhập từ bàn phím).
2. Tính tổng tiền lương công ty phải trả cho toàn bộ nhân viên trong một tháng.

```python
class Employee():
  def _init_(self,id,name,dob,address,salary,start):
    self.id = id
    self.name = name
    self.dob = dob
    self.address = address
    self.salary = salary
    self.start = start
  def input(self):
    self.id = int(input('Enter the id of employee: '))
    self.name = input('Enter name of employee: ')
    self.dob = input('Enter date of birth: ')
    self.address =  input('Enter current address of employee: ')
    self.salary = int(input('Enter basic salary of employee: '))
    self.start = input('Enter first day of work: ')

class Manager(Employee): 
  def _init_(self):
    Employee._init_(self)
  def calc_wage(self):
    return self.salary*4000000

class Programmer(Employee): 
  def _init_(self,ot):
    Employee._init_(self)
    self.ot = ot
  def input(self):
    Employee.input(self)
    self.ot = int(input('Enter money earned from overtime: '))
  def calc_wage(self):
    return self.salary*4000000 + self.ot

class Designer(Employee):
  def _init_(self,bonus):
    Employee._init_(self)
    self.bonus = bonus
  def input(self):
    Employee.input(self)
    self.bonus = int(input('Enter money earned from bonus: '))
  def calc_wage(self):
    return self.salary*4000000 + self.bonus

class Tester(Employee):
  def _init_(self,error):
    Employee._init_(self)
    self.error = error
  def input(self):
    Employee.input(self)
    self.error = int(input('Enter the error found: '))
  def calc_wage(self):
    return self.salary*4000000 + self.error*200000
```

```python
n = int(input('Enter the number of employee: '))
print('-----Provide some necessary information of employee-----')
l = []
for i in range(n):
  job = [Manager(),Programmer(),Designer(),Tester()]
  print('\nPlease choose: 0 - Manager, 1 - Programmer, 2 - Designer, 3 - Tester')
  j = int(input('Enter the type of job: '))
  a = job[j]
  a.input()
  l.append(a)
```

```python
paid = 0
for i in l:
  d = i.calc_wage()
  paid += d
  print(i.id,i.name,i.dob,i.address,i.salary,i.start,d)
```

```python
class Guest():
  def _init_(self,name,id):
    self.name = name
    self.id = id
  def input(self):
    self.name = input('Enter guest name: ')
    self.id = input('Enter guest id: ')

class Premium(Guest):
  def _init_(self,basic_cost,Class):
    Guest._init_(self)
    self.basic_cost = basic_cost
    self.Class = Class
  basic_cost = 1000
  Class = 0
  def input(self):
    Guest.input(self)
    l = list(map(int,input('Choose the services: ').split()))
    if l[0]:
      self.Class = int(input('How many class that you want to enroll: '))
  def cost(self):
    return self.basic_cost

class Basic(Guest):
  def _init_(self,basic_cost,Class,sup):
    Guest._init_(self)
    self.basic_cost = basic_cost
    self.Class = Class
    self.sup = sup
  basic_cost = 500
  def input(self):
    Guest.input(self)
    l = list(map(int,input('Choose the services: ').split())) 
    self.Class = 0
    self.sup = l[2]
    if l[0]:
      self.Class = int(input('How many class that you want to enroll: '))
    if l[1]:
      print('Sorry, you can not register steambath service because your package does not has this service.')
  def cost(self):
    return self.basic_cost + self.Class*100 + self.sup*100

class Non_member(Guest):
  def _init_(self,basic_cost,sup):
    Guest._init_(self)
    self.basic_cost = basic_cost
    self.sup = sup
  basic_cost = 200
  def input(self):
    Guest.input(self)
    l = list(map(int,input('Choose the services: ').split())) 
    self.sup = l[2]
    if l[0]:
      print('Sorry, you can not register class service because your package does not have this service.')
    if l[1]:
      print('Sorry, you can not register steambath service because your package does not have this service.')
  def cost(self):
    return self.basic_cost + self.sup*200
```

```python
a = Premium()
a.input()
print(a.name,a.id,a.basic_cost,a.Class)
a.cost()
```

```python
a = Basic()
a.input()
print(a.name,a.id,a.basic_cost,a.Class,a.sup)
print(a.cost())
```

```python
l = []
n = int(input('Enter the number of guest: '))
print('-----Provide necessary information of guest-----')
print('There are 3 pack: 0 - Premium, 1 - Basic, 2 - Non-member')
print('Please choose 1 of 3 packs above!')
print('We have 3 main services: class, steambath and PT support, if you want to select, please choose 1 to activate each service and if no just enter 0.')
print('Example: 1 1 1 (activate 3 service, and its work respectively)')
for i in range(n):
  print('\n')
  pack = [Premium(),Basic(),Non_member()]
  p = int(input('Enter the pack want to use: '))
  t = pack[p]
  t.input()
  l.append(t)
```

```python
for i in l:
  print(i.name, i.id, i.cost())
```

```python
def money(a):
  return a.cost()
customer_consult = []
l.sort(key=money,reverse=True)
for i in l:
  if i.cost()>=1000:
    customer_consult.append(i)
  else: break
print('Customers who are potential and can be consulted:')
for i in customer_consult:
  print('Name ' + i.name + ', id = ' + str(i.id) + ', used: '+str(i.cost()) + '000VND')
```
