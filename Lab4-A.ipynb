# Bài tập 1. Phép toán cơ bản với ma trận

```python
import numpy as np
```

Câu 1

```python
A = np.array([[2, 1, -1], [0, 1, -4]])
B = np.array([[-2, 1, 0], [-3, 2, 2]])
print(3*A+2*B)
```

Câu 2

```python
A = np.array([[2,-5,1],[3,0,-4]])
B = np.array([[1,-2,-3],[0,-1,5]])
C = np.array([[0,1,-2],[1,-1,-1]])
print(3*A+4*B-2*C)
```

Câu 3

```python
A = np.array([[0,0,1],
             [1,1,2],
             [2,2,3],
             [3,3,4]])
B = np.array([[-1,-1],[2,2],[1,1]])
C = np.array([[4],[1]])
print(A@B@C)
```

Câu 4

```python
A = np.array([[1,3,5],[6,-7,-8]])
AT = np.transpose(A)
print('A^T:\n',AT)
print('(A^T)^T:\n',np.transpose(AT))
print('A(A^T):\n',A@AT)
print('(A^T)A:\n',AT@A)
```

Câu 5

```python
A = np.array([[1,2],[3,-4]])
B = np.array([[5,0],[-6,7]])
print('(AB)^T:\n',np.transpose(A@B))
print('(A^T)(B^T):\n',np.transpose(B@A)) #= (np.transpose(A))@(np.transpose(B)))  
print('(B^T)(A^T):\n',np.transpose(A@B)) #= (np.transpose(B))@(np.transpose(A)))
```

Câu 6

```python
A = np.array([[2,0,0],[3,-1,0],[4,-2,5]])
B = np.array([[2,0],[1,-3],[4,5]])
C = A@B
print('AB:\n',C)
print('(A^T)A:\n',A@(np.transpose(A)))
print('AB-3B:\n',C-3*B)
```

Câu 7

```python
A = np.array([[1,2,3,4],
              [2,5,4,7],
              [3,7,8,12],
              [4,8,14,19]])
print('A^-1:\n',np.linalg.inv(A))
```

Câu 8

```python
from numpy.linalg import matrix_power
A = np.array([[0,1,0],[0,0,1],[0,0,0]])
AT = np.transpose(A)
print('A(A^T):\n',A@AT)
print('(A^T)A:\n',AT@A)
print('A^2:\n',matrix_power(A,2))
print('A^3:\n',matrix_power(A,3))
```

```python
from numpy.linalg import matrix_power
A = [[1,-3],[2,5]]
print(matrix_power(A,2))
print(np.square(A))
```

Câu 9

```python
A = np.array([[1,0,-1],[-2,1,4],[3,2,2]])
B = np.array([[4,-1,2],[-6,7,-3],[1,0,-8]])
C = np.array([[3,-2,-1],[-4,0,9],[2,-7,5]])
print('AB-(A^T)C:\n',A@B-(np.transpose(A)@C))
```

Câu 10

```python
A = np.array([[1,3],[-1,2],[3,4]])
B = np.array([[0,1],[3,2],[-2,3]])
C = np.array([[2,-3],[1,2],[4,-1]])
print('A+(B+C):\n',A+B+C)
print('(A+B)+C:\n',A+B+C)
print('3A:\n',3*A)
print('A^T:\n',np.transpose(A))
print('B^T:\n',np.transpose(B))
print('C^T:\n',np.transpose(C))
```

# Bài tập 2. Đa thức ma trận

Câu 1

```python
A = np.array([[1,-2,3],[2,-3,1],[3,-5,2]])
f_a = 3*matrix_power(A,2) - 2*A + 5
print(f_a) 
```

Câu 2

```python
A = np.array([[2,2],[3,-1]])
f_a = matrix_power(A,3) - 3*matrix_power(A,2) - 2*A + 4
print(f_a)
```

Câu 3

```python
A = np.array([[1,1],[0,1]])
f_a = matrix_power(A,3) -3*A + 2*matrix_power(A,-1)
print(f_a)
```

Câu 4

```python
A = np.array([[0,1,1],[1,0,1],[1,1,0]])
f_a = matrix_power((matrix_power(A,2)-1),-1) - matrix_power((matrix_power(A,2)+1),-1)
print(f_a)
```

# Bài tập 3. Định thức của ma trận


Câu 1:

```python
A = np.array([[1+2**0.5,2-3**0.5],[2+3**0.5,1-2**0.5]])
print('det(A) =',round(np.linalg.det(A)))
```
