# Import thư viện

```python
from collections import Counter
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import statistics as stats
```

```python
def five_stats(a):
  print('Số trung bình: ',stats.mean(a))
  print('Độ lệch chuẩn: ',stats.stdev(a))
  print('Số trung vị: ',stats.median(a))
  print('Tứ phân vị: ',[round(q, 1) for q in stats.quantiles(a, n=4)])
  print('Mốt: ',stats.mode(a))
  print('Phương sai: ',stats.variance(a))
```

# Câu 22 

a. Tính hệ số tương quan giữa ứng suất pháp tuyến và độ kháng cắt?
Nêu nhận xét về mối tương quan giữa hai đại lượng này?

```python
uspt = [26.8,25.4,28.9,23.6,27.7,23.9,24.7,28.1,26.9,27.4,22.6,25.6]
dkc = [26.5,27.3,24.2,27.1,23.6,25.9,26.3,22.5,21.7,21.4,25.8,24.9]
print(len(uspt),len(dkc))
```

```python
stats.correlation(uspt,dkc)
```

Vì hệ số tương quan giữa ứng suất pháp tuyến và độ kháng cắt là số âm nên mối quan hệ giữa hai đại lượng là mối quan hệ nghịch biến hoặc tương quan âm.

b. Ước lượng đường hồi quy tuyến tính? Vẽ các điểm dữ liệu và đường
hồi quy trên cùng một đồ thị? Ước lượng độ kháng cắt ứng với ứng
suất pháp tuyến 24.5?

```python
from scipy import stats
res = stats.linregress(uspt,dkc)
print(f"R-squared: {res.rvalue**2:.6f}")
```

```python
print(type(uspt))
print(type(res.intercept),type(res.slope))
```

```python
plt.plot(uspt, dkc, 'o', label='original data')
plt.plot(uspt, res.intercept + res.slope*np.array(uspt), 'r', label='fitted line')
plt.legend()
plt.show()
```

Ước lượng độ kháng cắt với pháp tuyến ứng suất 24.5 là

```python
res.intercept + res.slope*24.5 
```

```python
dbc = np.array([2158.7,1678.15,2316,2061.3,2207.5,1708.3,1784.7,2575,2357.9,2277.7,2165.2,
                2399.55,1779.8,2336.75,1765.3,2052.5,2414.4,2200.5,2654.2,1753.7])
tdc = np.array([15.5,23.75,8,17,5,19,24,2.5,7.5,11,
                13,3.75,25,9.75,22,18,6,12.5,2,21.5])
print(len(dbc),len(tdc))
```
