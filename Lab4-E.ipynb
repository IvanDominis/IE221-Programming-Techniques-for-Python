# C√¢u 2.5: Gi·∫£i c√°c ph∆∞∆°ng tr√¨nh vi ph√¢n sau:

```python
from sympy.abc import x
y = Function('y')
eq = Eq(y(x).diff(x) - 2*x*(x**2+16)**0.5,0)
dsolve(eq)
```
